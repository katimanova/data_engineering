### Получим Markdown с помощью бибилиотеки Docling
Ссылочка - [Docling GitHub Repository](https://github.com/DS4SD/docling)

Это нужно для того, чтобы понять какими средствами лучше разбивать и структуризирвоать pdf документы. 

````python
from docling.document_converter import DocumentConverter
converter = DocumentConverter()
````

````python
source = "../../data/Base_Books/Волга/Книга 1/Книга 1.pdf" 

result = converter.convert(source)
print(result.document.export_to_markdown())
````

````python
# Экспортируем текст в Markdown
markdown_content = result.document.export_to_markdown()

# Записываем в файл
output_file = "./outputs/Волга_Книга_1.md"
with open(output_file, "w", encoding="utf-8") as file:
    file.write(markdown_content)

print(f"Результаты сохранены в {output_file}")
````
