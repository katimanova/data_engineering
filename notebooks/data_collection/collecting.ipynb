# Сбор данных

### Почему этот этап важен?
Сбор данных — это первостепенный этап, так как без данных невозможно выполнить дальнейшие шаги: анализ, подготовку и разработку. 


### Особенности сбора данных
В моем случае процесс сбора данных был осложнен следующими факторами:
- **Разнородность источников:** Каждое БВУ (Бассейновое водное управление) выкладывало отчеты отдельно на своих платформах.

### Объем и структура данных
На данный момент все данные хранятся в архиве. Вот основные характеристики:

1. **Объем данных:**
   - Всего 80 бассейнов.
   - По каждому бассейну минимум 6 книг.
   - Каждая книга описывает отдельные аспекты + приложения к ним.

2. **Структура отчетов:**
   - Множество таблиц, графиков и приложений.
   - Объемы отдельных отчетов варьируются от **40 до 350 страниц**.

3. **Проблемы с размещением данных:**
   - Из-за большого объема архив невозможно прикрепить в платформы вроде GitHub.

### Сбор данных
Этот этап является первостепенным - получить данные, а уже дальше анализировать их, подготваливать и так далее. В моем случае - сбор данных осложнен тем фактом, что все каждое БВУ (бассейновое водное управление) выкладывало отдельно у себя все отчеты. Но так как регламент написания составляло Донское БВУ, то в силу проверки качества все отчеты передавались нам напрямую (почтой, флешкой ... этот процесс длится около 10 лет как мне сказали).

На данные момент все хранится в архиве (сюда я его прикрепить не могу в силу большого объема. Для понимания масштаба - всего 80 бассейнов, по кажлому минимум написано 6 книг, где каждая отвечает за определнные аспект + приложения практически к каждой книге. Также в отчетах много таблц, графиков и объемы от 40 страниц до 350 -> что делает архивы вестма большим и громоздким. Гитхаб просто не пропустит)


###  Данные есть. Они в архиве

#### Работа с архивом
Архив с данными находится локально на компьютере. На данный момент, согласно ТЗ, в рассмотрении находятся данные по следующим 8 бассейнам:

- **Волга**
- **Дон**
- **Кубань**
- **Обь**
- **Печора**
- **Сура**
- **Терек**
- **Урал**

### Почему нельзя самой собрать все отчеты?
Как я и сказала ранее, процесс сбора пропущен. Однако по заданию от курса я также показала, почему парсинг с сайта не делает процесс автоматизированным в альтернативу готового архива.

#### Проблемы с форматом данных:
Как видно, при переходе от одного БВУ к другому:
- Формат данных изменяется.
- Невозможно предугадать, где это будет архив, ссылка для скачивания, или информация о файлах в дополнительном документе (.docx).


#### Вывод:
Использование готового архива позволяет избежать сложностей с автоматизацией, так как парсинг таких разнородных источников потребует значительных временных затрат, но по эффективности не принесет пользы.
На данный момент данные для работы хранятся по пути /Base_Books/
